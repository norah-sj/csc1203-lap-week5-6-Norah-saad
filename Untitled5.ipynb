```python
import math

# Function to calculate area
def calculate_area(radius):
    return math.pi * radius ** 2

# Lambda to calculate circumference
circumference = lambda r: 2 * math.pi * r

# User input
radius = float(input("Enter the radius of the circle: "))

area = calculate_area(radius)
circ = circumference(radius)

print("Area:", round(area, 2))
print("Circumference:", round(circ, 2))
```

```python
import re

# Function to sum numbers
def sum_numbers(numbers):
    return sum(numbers)

# User input
text = input("Enter a string: ")

# Find all numbers in string
numbers = re.findall(r'\d+', text)
numbers = list(map(int, numbers))

total = sum_numbers(numbers)

# Lambda to double the sum
double = lambda x: x * 2

print("Numbers found:", numbers)
print("Sum:", total)
print("Doubled value:", double(total))
```

```python
import math

# Function to calculate squares using lambda
def calculate_squares(numbers):
    square = lambda x: x ** 2
    return list(map(square, numbers))

# User input
nums = input("Enter numbers separated by commas: ")
numbers = list(map(float, nums.split(",")))

squares = calculate_squares(numbers)

max_square = max(squares)
sqrt_sum_squares = math.sqrt(sum(squares))

print("Squares:", squares)
print("Maximum square:", max_square)
print("Square root of sum of squares:", sqrt_sum_squares)
```

```python
import re
import math

# Password check
password = input("Enter password: ")

upper = re.search(r'[A-Z]', password)
lower = re.search(r'[a-z]', password)
number = re.search(r'\d', password)

if upper and lower and number:
    strength = "Strong"
else:
    strength = "Weak"

# Scores input
scores = list(map(float, input("Enter 5 scores separated by spaces: ").split()))

# Function to calculate average using lambda
def calculate_average(scores):
    avg = lambda lst: sum(lst) / len(lst)
    return avg(scores)

average = calculate_average(scores)
rounded_avg = math.ceil(average)

print("Password is:", strength)
print("Rounded Average Score:", rounded_avg)
```

```python
import re
import math

# Function to count total vowels
def count_vowels(text):
    vowels = re.findall(r'[aeiouAEIOU]', text)
    return len(vowels)

# User input
paragraph = input("Enter a paragraph: ")

# Find words starting with vowel
vowel_words = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)

# Lambda to calculate length of each vowel word
word_length = lambda word: len(word)

lengths = list(map(word_length, vowel_words))

# Calculate average word length
if lengths:
    average_length = round(sum(lengths) / len(lengths), 2)
else:
    average_length = 0

print("Words starting with vowels:", vowel_words)
print("Total vowels in paragraph:", count_vowels(paragraph))
print("Average vowel-word length:", average_length)
```
